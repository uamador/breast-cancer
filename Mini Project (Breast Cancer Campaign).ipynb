# 4 Preprocessing
Perform any needed pre-processing on the chosen features, including:
* Scaling
* Encoding
* Dealing with NaN values
* Normalization

Note:
You need to output the result of your pre-processing to an output CSV called “data_refined.csv.”

```python
# Need to encode the diagnois column. This is also the target column Y

from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()
df['diagnosis'] = enc.fit_transform(df['diagnosis'])
```

```python
df.head()
```

```python
# I need to normalize the columns since the ranges between columns vary greatly. I will use the MinMaxScaler to make all colun values between 0 and 1

from sklearn.preprocessing import MinMaxScaler

# The diagnosis coluns is lable encoded. It needs to be excluded from the normalization.

X = df.drop('diagnosis', axis=1)

y = df['diagnosis']

scaler = MinMaxScaler()

X_normalized = scaler.fit_transform(X)

# Need to create new dataframe
df_normalized = pd.DataFrame(X_normalized, columns=X.columns)

# add the y (diagnosis) column
df_normalized = pd.concat([y, df_normalized], axis=1)

# overwrite the old dataframe
df = df_normalized
df.to_csv('data_refined.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
correlation_matrix = df.drop('diagnosis', axis=1).corr()


# Set up the matplotlib figure
plt.subplots(figsize=(15,12))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(correlation_matrix)

plt.show()
```

```python
sns.scatterplot(x='area_mean', y='smoothness_mean', hue='diagnosis', data=df)
plt.show()
```

```python
# It looks like area_mean of > 0.2 strongly correlate to a malignant diagnosis
```

```python
sns.boxplot(x='diagnosis', y='texture_mean', data=df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6)) 
sns.barplot(x='diagnosis', y='radius_mean', data=df)

plt.title('Average Radius for Benign vs Malignant Diagnoses')
plt.xlabel('Diagnosis')
plt.ylabel('Mean Radius')

plt.show()
```

```python
#bigger radius mean the more likely to be a malignant diagnosis
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6)) 
sns.violinplot(x='diagnosis', y='radius_mean', data=df, palette="muted")

plt.title('Distribution of Mean Radius for Benign vs Malignant Diagnoses')
plt.xlabel('Diagnosis')
plt.ylabel('Mean Radius')

plt.show()
```
